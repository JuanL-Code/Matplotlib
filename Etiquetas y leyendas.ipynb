```python
import numpy as np
import seaborn as sns  
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('base_datos_2008.csv.bz2', nrows = 100000)

data = np.unique(df.DayOfWeek, return_counts = True)
labs=["Lun", "Mar", "Mie", "Jue", "Vie", "Sab", "Dom"]
data
```

```python
plt.pie(x = data[1], labels=labs, radius=1.5, colors=["Red", "Green", "Orange", "Blue", "Gray", "Pink", "Black"], startangle=90)
plt.show()
```

```python
#Explode, sacar porcion 7
from tkinter.ttk import LabeledScale


plt.pie(x = data[1], 
    #labels=labs, 
    radius=.7, 
    colors=sns.color_palette("hls",7),
    explode=(0,0,0,0.2,0,0,0.1),
    startangle=90,
    autopct="%1.1f%%")
plt.legend(loc="upper right", labels = labs)
plt.show()
```

```python
plt =  sns.barplot(x=labs, y=data[1])
plt.set(xlabel="Dia de la semana", ylabel="numero de vuelos")
```
