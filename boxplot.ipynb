```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('base_datos_2008.csv.bz2', nrows = 100000)
```

```python
df.dropna(inplace=True, subset = ["ArrDelay", "DepDelay", "Distance"])
```

```python
sns.displot(df["Distance"])
```

```python
sns.displot(df["Distance"], kde = False, bins=50)
```

```python
sns.kdeplot(df["ArrDelay"])
sns.kdeplot(df["DepDelay"])
plt.xlim([-30, 300])
```
