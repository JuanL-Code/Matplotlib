```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(0)
```

```python
df = pd.read_csv('base_datos_2008.csv.bz2', nrows = 100000)
df = df.sample(frac=1). head(100)
```

```python
plt.scatter(x=df.DayofMonth, y=df.ArrDelay, s=df.Distance)
```

```python
#6 semana, 7 Fines
plt.scatter(x=df.DayofMonth, y=df.ArrDelay, s=df.Distance, alpha=.3, c = df.DayOfWeek.isin([6,7]))
plt.title("Retrasos en EEUU")
#Vuelos a tiempo, reducir el recuadro a esos
plt.ylim([0,150])
#Limitar intervalos a mostrar
plt.xticks([0,15,30])
plt.text(x=28,y=120, s="mi vuelo")
```
